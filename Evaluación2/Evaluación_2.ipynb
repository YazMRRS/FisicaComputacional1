```python
#Importamos bibliotecas 
import numpy as np
from scipy.integrate import odeint, solve_ivp
import matplotlib.pyplot as plt
import scipy.linalg as la
from scipy.integrate import odeint
```

```python
#Definimos la función del oscilador amortiguado
def oa(y,t,C,z,m,f0,o):
  posición, velocidad = y
  dydt = [velocidad, f0*np.sin(o*t)/m-2.0*z*C*velocidad-C**2*posición]
  return dydt 
```

```python
#Problema 1

#Definimos las constantes
m=1
f0=0
k=9
C=np.sqrt(k/m)
o=0

#Definimos los valores de z que son de nuestro interes
z1=0
z2=0.1
z3=0.2
z4=0.4
z5=0.7
z6=1
z7=2

t = np.linspace(0,3*np.pi,500)
y0 = [-1.0,0.0]

#Obtenemos las soluciones correspondientes con ODEINT
solución1 = odeint(oa,y0,t,args=(C,z1,m,f0,o))
solución2 = odeint(oa,y0,t,args=(C,z2,m,f0,o))
solución3 = odeint(oa,y0,t,args=(C,z3,m,f0,o))
solución4 = odeint(oa,y0,t,args=(C,z4,m,f0,o))
solución5 = odeint(oa,y0,t,args=(C,z5,m,f0,o))
solución6 = odeint(oa,y0,t,args=(C,z6,m,f0,o))
solución7 = odeint(oa,y0,t,args=(C,z7,m,f0,o))

plt.figure(figsize=(10,5))
plt.title('Oscilador amortiguado sin forzamiento-Posición contra tiempo')
plt.xlabel('t')
plt.ylabel('x(t)')

plt.plot(t, solución1[:,0], label='z=0')
plt.plot(t, solución2[:,0], label='z=0.1')
plt.plot(t, solución3[:,0], label='z=0.2')
plt.plot(t, solución4[:,0], label='z=0.4')
plt.plot(t, solución5[:,0], label='z=0.7')
plt.plot(t, solución6[:,0], label='z=1')
plt.plot(t, solución7[:,0], label='z=2')

plt.legend(loc='best')
plt.show()
```
