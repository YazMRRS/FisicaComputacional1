```python
#Importamos bibliotecas 
import numpy as np
import scipy.linalg as la
from numpy.linalg import matrix_power as mpow
import matplotlib.pyplot as plt
from scipy.integrate import odeint, solve_ivp
from mpl_toolkits import mplot3d
from scipy import integrate, linalg
import pylab as plt

%matplotlib inline
```

```python
def parametro(M):
  '''
  Patámetro
  ----------
  M: matriz a la que se le va a sacar los valores 

  Regresa
  -------
  Eigenvalores
  Delta
  Traza
  Determinante
  '''

  eig_vals, eig_vects = la.eig(M)
  tr=np.trace(M)
  det = la.det(M)
  delta = (tr)**2-(4*det)
  print('Eigenvectores', eig_vects)
  print()
  print('Eigenvalores', eig_vals)
  print()
  print('Traza de A1:', tr)
  print()
  print('Determinante de A1:', det)
  print()
  print('Delta de A1:', delta)
  print()
```

```python
#Ejercicio1

#Determinamos la matriz de la primera ecuación
A1=np.array([[0,1],[-1,0]])
parametro(A1)

#Resolvemos la ecuación diferencial, para esto, definimos un vector X tal que X[0]=x, X[1]=y
def dXdt(X, t=1):
  return np.array([X[1], -X[0]])
#Definimos los puntos criticos 
Xf0 = np.array([0.0,0.0])
all(dXdt(Xf0) == np.zeros(2))

#Solución numerica
t=np.linspace(0.0,2*np.pi, 51)
X0=np.array([1.0,1.0])
#Aplicamos condiciones inciales
X, infodict=integrate.odeint(dXdt,X0,t,full_output=True)
infodict['message']

x,y=X.T

f1=plt.figure();
plt.figure(figsize=(10,5));
plt.plot(t,x,'g.-', label='x(t)');
plt.plot(t,y,'r.-', label='y(t)');
plt.grid();
plt.legend(loc='best');
plt.xlabel('Tiempo');
plt.ylabel('Soluciones');
plt.title('Soluciones para x(t) & y(t)');
plt.show

#Calculamos el plano fase
f1=plt.figure();
plt.plot(x,y,'red');
plt.xlabel('x');
plt.ylabel('y');
plt.grid();
plt.title('Plano fase - Solución para X(t)');
plt.show()
```
